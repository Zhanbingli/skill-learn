```python
df = pd.DataFrame({
    "user": ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C'],
    "value": [5, 15, 25, 10, 20, 30, 5, 15],
    "data": pd.to_datetime(["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04",
                              "2023-01-05", "2023-01-06", "2023-01-07", "2023-01-08"])

})

print(df.head())
print(df.info())
print(df.describe())
```

```python
print(f'value :{df["value"]}')
print(df.loc[0])
print(df.iloc[1:4])
```

```python
print(df["value"] > 10)
```

```python
print(df[df["value"] > 10])
print("="*20)
print(df[df["value"] > 10]["user"].iloc[1:4])
```

```python
df["value_norm"] = (df["value"] - df["value"].mean()) / df["value"].std()
print(df)
```

```python
res = df.groupby("user", as_index=False)["value"].mean().sort_values(by="value", ascending=False)
print(res)
```

```python
df2 = pd.DataFrame({
    "name": ['Alice', 'Bob', None, 'Alice', 'Col'],
    "score": [90, 80, 70, None, None],
    "tag": ['A', 'B', 'C', 'A', None]

})

print("before:\n" ,df2)
```

```python
print("isnull:\n", df2.isnull())
```

```python
df2_filled = df2.fillna({"score": df2["score"].mean(), "name": "Unknown"})
print("after fillna:\n", df2_filled)
```

```python
def abc(score):
    if score >= 85:
        return 'A'
    elif score >= 75:
        return "B"
    else:
        return "C"
df2_filled['grade'] = df2_filled['score'].apply(abc)
print("after apply:\n", df2_filled)
```

```python
print('drop duplicates:\n', df2_filled.drop_duplicates("name")) #这里需要传入参数
```

```python
left = pd.DataFrame({
    'id': [1, 2, 3],
    'value_left': ['A', 'B', 'C']
})
right = pd.DataFrame({
    'id': [2, 3, 4],
    'value_right': ['D', 'E', 'F']
})
print("merge inner:\n", left.merge(right, on='id', how='inner'))

sales = pd.DataFrame({
    'product': ['A', 'B', 'C', 'A'],
    'day': ['2023-01-01', '2023-01-01', '2023-01-01', '2023-01-02'],
    'sales': [100, 150, 200, 130],
    'amout': [1000, 2000, 3000, 1500]})
pivot = sales.pivot_table(index='product', values=['sales', 'amout'], aggfunc='sum', fill_value=0, columns=['day'])
print("pivot table:\n", pivot)
```
