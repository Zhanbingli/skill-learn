```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# 生成医学数据示例
np.random.seed(42)
patients = pd.DataFrame({
    'age': np.random.normal(45, 15, 1000),
    'systolic_bp': np.random.normal(120, 20, 1000),
    'diastolic_bp': np.random.normal(80, 10, 1000),
    'cholesterol': np.random.normal(200, 40, 1000)
})

# 描述性统计
print("描述性统计:")
print(patients.describe())

# 正态性检验
statistic, p_value = stats.normaltest(patients['systolic_bp'])
print(f"血压正态性检验: p = {p_value:.4f}")

# 相关性分析
correlation = patients.corr()
print("相关系数矩阵:")
print(correlation)
```
